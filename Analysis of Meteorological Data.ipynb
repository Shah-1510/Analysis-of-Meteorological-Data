# Dropping Columns That Are Not Relevant To The Analysis

# from the dataset, it seems that the "Summary" column provides the necessary information about the weather condition and the #"loud cover" column has 0 values throughout the column data so it defines no meaning to the data.

# So, I will be dropping these columns from the dataset.

```python
data.drop(['Loud Cover','Daily Summary'],axis=1,inplace=True)
data
```

# Resampling the Data With Respect to "Months".

```python
data[['Formatted Date']]
```

```python
# This is the breakdown of the data w.r.t months.
data['Formatted Date']=pd.to_datetime(data['Formatted Date'],utc=True)
data
df = data.set_index('Formatted Date')
df = df.resample('M').mean()
df
```

# Simple Comparision of Apparent Temperature vs. Humidity

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
g = sns.relplot(x='Apparent Temperature (C)', y='Humidity', data=data, kind="scatter", color="green")
g.fig.suptitle("Apparent Temperature (C) vs. Humidity",y=1.1)
plt.show()

from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

# Comparision of Apparent Temperature vs. Humidity with respect to "Summary" of the weather on a given day using Pair Plots.

```python
g = sns.pairplot(data, vars=['Humidity', 'Apparent Temperature (C)'], hue='Summary')
g.fig.suptitle('Apparent Temperature (C) vs. Humidity w.r.t Summary', y=1.2)
plt.show()
```

# Comparision of Apparent Temperature vs. Humidity with respect to "Summary" of 
# the weather on a given day using Relational Plots with Columns separated as "Summary".

```python
plt.figure(figsize=(16,6))
g = sns.relplot(x='Apparent Temperature (C)',y='Humidity', data=data,color='green',hue='Summary',col='Summary',col_wrap=3)
g.fig.suptitle('Apparent Temperature (C) vs. Humidity w.r.t Summary',y=1.005)
plt.show()
```

# 2D Scatter plot with color coding provided for each summary type

# As we have seen 

# There are very few outlier.
# Mostly Weather is Clear or Partly Cloudy/Rain in Finland.
# Only few days there has a Light Rain or Dry or Dangerously Windy and Partly Cloudy.

```python
sns.set_style("darkgrid")
sns.FacetGrid(data, hue="Summary", height=10).map(plt.scatter, "Apparent Temperature (C)", "Humidity").add_legend()
plt.title("2D Scatter Plot with Color Coding for each Summary type")
plt.show()
```
